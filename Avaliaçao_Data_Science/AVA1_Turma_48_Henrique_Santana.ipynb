*** Questão número 1 ***

Aqui está uma breve explicação do que cada termo significa:

Região: Uma área geográfica definida por características comuns, como localização, clima, cultura, etc.

Estado: Uma divisão administrativa dentro de um país, geralmente com seu próprio governo e leis.

Município: Uma divisão administrativa dentro de um estado, geralmente correspondendo a uma cidade ou vila.

Código_mun: Um identificador numérico único atribuído a cada município para fins de organização e referência.

Capital: A cidade que serve como sede do governo de um estado ou país.

Porte: O tamanho ou magnitude de algo, frequentemente usado para descrever o tamanho de uma cidade ou município.

População: O número de pessoas que vivem em uma determinada área.

Área: A extensão de terra ocupada por uma região, município ou país, geralmente medida em quilômetros quadrados.

Densidade_dem: A densidade populacional de uma área, calculada como o número de pessoas por unidade de área.

PIB: Produto Interno Bruto, uma medida do valor total de todos os bens e serviços produzidos
dentro das fronteiras de um país ou região em um determinado período de tempo.

Gasto_pc_educacao: O gasto médio por pessoa em educação em uma determinada área.

Exp_anos_estudo: Expectativa de anos de estudo, uma estimativa do número médio de anos de escolaridade que uma pessoa pode esperar receber.

Nota_ciencias: Pontuação média em ciências em um determinado contexto educacional.

Nota_humanas: Pontuação média em disciplinas de humanas (como história, geografia, etc.) em um determinado contexto educacional.

Nota_linguagem: Pontuação média em disciplinas de linguagem (como português, inglês, etc.) em um determinado contexto educacional.

Nota_mat: Pontuação média em matemática em um determinado contexto educacional.

Nota_redacao: Pontuação média em redação em um determinado contexto educacional.

IDHM: Índice de Desenvolvimento Humano Municipal, uma medida composta que avalia o desenvolvimento humano em uma área com base em indicadores como saúde, educação e renda.

Ranking_igm: Ranking de Índice de Governança Municipal, uma medida que classifica os municípios com base em sua governança e eficiência administrativa.

Indice_governanca: Um índice que mede a qualidade da governança em uma determinada área, levando em consideração vários aspectos, como transparência, eficácia administrativa e participação cidadã.

Uma das formas de tratar os dados faltantes em cada uma das variaveis é substituir os dados faltantes pela média, mediana ou moda da coluna escolhida, isso é util porque preserva o tamanho do conjunto de dados, melhora a performance do modelo, e facilita a análise exploratório ( visto que não mudamos muito os resultados finais por causa dos missing values ). Nesse conjunto de dados, a melhor forma seria substituir os "missing values" pela média, pois não mudariamos muito o calculo final do modelo, pois não temos muitos valores faltantes na maioria das variaveis e não seria necessário excluir a coluna.

```python
# A coluna escolhida foi a "servidores" que tem 337 missing values, trocarei o
# NaN pela mediana como solicitado.

median_serv = mediana = df['servidores'].median()
print(median_serv)
```

```python
# Substituindo os missing values da coluna servidores pela mediana
# que foi anteriormente definida como "median_serv".

df['servidores'].fillna(median_serv, inplace=True)
```

Hipótese: Quanto maior o pib da região, maior será o gasto com a educação, consequentemente maior o idh, logo a nota de redação da região será maior.

```python
df.head()
```

```python
#hipótese :
#regiao, pib > gasto com educaçao > idhm > nota redação
# Criando um dataframe menor chamado df1 com as variaveis regiao,
# gasto_pc_educacao, idhm, nota_redacao

df1 = df[["regiao", "pib", "gasto_pc_educacao","idhm", "nota_redacao"]]

# Número de linhas e colunas do dataframe
df1.shape
```

```python
#podemos ver que temos 5 regiões e a que aparece mais vezes é a NORDESTE
# que temos 342 numeros de idh diferente, e 4447 notas de redações diferentes
# temos 5570 dados de região, 4986 dados de idh e 5557 dados de nota de redação.

df1.describe(include="object")
```

```python
# Para podermos consolidar nossa hipotese irei pegar o valor do PIB de cada região e verificicar o valor total de cada um

#Realizando a soma do valor do PIB nas regiões selecionadas
soma_total_Pib_Sudeste = df1.loc[df1['regiao'] == "SUDESTE", "pib"].sum()
soma_total_Pib_Sul = df1.loc[df1['regiao'] == "SUL", "pib"].sum()

df1 = df1[["regiao", "pib", "gasto_pc_educacao","idhm", "nota_redacao"]].loc[df1['regiao'].isin(['SUDESTE', "SUL"])]

print("Pib total da região sudeste",soma_total_Pib_Sudeste)
print("PIB total da região SUL", soma_total_Pib_Sul)


print("Com Esse resultado podemos perceber que a região SUL tem um PIB maior que a região SUDESTE, sabendo disso podemos prever que o gasto com a educação da região SUL é maior do que na região SUDESTE\n")

# Valores a serem plotados no grafico
somas_pib = [soma_total_Pib_Sudeste, soma_total_Pib_Sul]
regioes = ['SUDESTE', 'SUL']

# Criando o gráfico de colunas
plt.bar(regioes, somas_pib, color=['blue', 'green'])

# Adicionando título e rótulos aos eixos
plt.title('PIB por Região')
plt.xlabel('Região')
plt.ylabel('PIB Total')

# Mostrando o gráfico
plt.show()
```

```python

#Realizando a Soma do valor Total gasto em educação
soma_total_Edu_Sudeste = df1.loc[df1['regiao'] == "SUDESTE", "gasto_pc_educacao"].sum()
soma_total_Edu_Sul = df1.loc[df1['regiao'] == "SUL", "gasto_pc_educacao"].sum()
print("Nessa comparação podemos verificar a compro.vação de que a região com maior PIB consegue investir mais em educação para a sua população.\n")

# Valores a serem plotados no grafico

regioes = ['SUDESTE', 'SUL']
gasto_educacao_total = [soma_total_Edu_Sudeste, soma_total_Edu_Sul]

# Criando o gráfico de barras
plt.bar(regioes, gasto_educacao_total, color=['blue', 'green'])



# Adicionando título e rótulos aos eixos
plt.title('Comparação do Gasto com Educação entre as Regiões Sudeste e Sul')
plt.xlabel('Região')
plt.ylabel('Gasto Total com Educação (em bilhões de reais)')

# Mostrando o gráfico
plt.show()
```

```python
# Convertendo os valores para números antes de calcular a média
idhm_sudeste = df1.loc[df1['regiao'] == "SUDESTE", "idhm"].astype(float).mean()
idhm_sul = df1.loc[df1['regiao'] == "SUL", "idhm"].astype(float).mean()

# Valores a serem plotados no gráfico
regioes = ['SUDESTE', 'SUL']
idhm = [idhm_sudeste, idhm_sul]

# Criando o gráfico de barras
plt.bar(regioes, idhm, color=['blue', 'green'])

# Adicionando título e rótulos aos eixos
plt.title('Comparação do IDHM entre as Regiões Sudeste e Sul')
plt.xlabel('Região')
plt.ylabel('IDHM')

# Mostrando o gráfico
plt.show()
```
